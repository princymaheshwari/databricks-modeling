```python
from pyspark.sql.functions import regexp_replace

df = spark.table("workspace.bronze.univital_seed_data")
```

```python
df = df.withColumn(
    "visit_category",
    regexp_replace("visit_category", "â€“", "-")
).withColumn(
    "age_group",
    regexp_replace("age_group", "â€“", "-")
)
```

```python
plans = df.filter("record_type = 'plan'")
util_priors = df.filter("record_type = 'utilization_prior'")
cost_priors = df.filter("record_type = 'cost_prior'")
er_priors = df.filter("record_type = 'er_visits_prior'")
subsidy_params = df.filter("record_type = 'subsidy_params'")
users = df.filter("record_type = 'synthetic_user'")
```

```python
plans.write.format("delta").mode("overwrite").saveAsTable("workspace.bronze.plans")
util_priors.write.format("delta").mode("overwrite").saveAsTable("workspace.bronze.utilization_priors")
cost_priors.write.format("delta").mode("overwrite").saveAsTable("workspace.bronze.cost_priors")
er_priors.write.format("delta").mode("overwrite").saveAsTable("workspace.bronze.er_priors")
subsidy_params.write.format("delta").mode("overwrite").saveAsTable("workspace.bronze.subsidy_params")
users.write.format("delta").mode("overwrite").saveAsTable("workspace.bronze.users")
```
