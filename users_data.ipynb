```python
from pyspark.sql.functions import col, lit

users = spark.table("workspace.bronze.users")

demo_user_ids = [
    "bd47636a-85a0-4427-b1a2-5d8bcb5d6255",
    "633205b2-6a1e-4be0-ad19-cd275900cad8",
    "cac71467-d073-49bd-aa4b-bc537fef23ef",
]

demo_users = (
    users
    .filter(col("user_id").isin(demo_user_ids))
    .select(
        "user_id",
        col("annual_income").cast("double").alias("annual_income"),
        col("age").cast("int").alias("age"),
        col("medication_count").cast("int").alias("medication_count"),
        col("expected_er_visits").cast("double").alias("expected_er_visits"),
        col("therapy_frequency").cast("double").alias("therapy_frequency"),
    )
    .withColumn("demo_county", lit("Fulton"))
    .withColumn("demo_label",
        col("user_id")  # optional; you can replace with nicer labels later
    )
)

demo_users.write.format("delta").mode("overwrite").saveAsTable("workspace.bronze.demo_users")
```
